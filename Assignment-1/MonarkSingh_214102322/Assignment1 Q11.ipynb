```python
# Assignment 1 Ques 11

import csv
import numpy as np

rows=[]
csvfile=open('Random_Point.csv', 'r')

csvreader = csv.reader(csvfile)

# Reading the CSV file 'Random_Point.csv' to a matrix

for row in csvreader:
    rows.append(row)

for row in rows: 
    for col in row:
        print(col)

# Closing the file
csvfile.close()
```
