```python
# Assignment 1 Ques 14 

import numpy as np
from matplotlib import pyplot as plt
from math import pi

num_samples = 100

h=5     # x-coordinate of the center
k=-5    # y-coordinate of the center
a=5     # Semi major axis
b=2.5   # Semi minor axis

# Defining angle from 0 to 2*pi
theta = np.linspace(0, 2*pi, 100)

# Plotting using parametric eqn of Ellipse: x=h+a*cos(theta) , y=k+b*sin(theta)
plt.plot(h+a*np.cos(theta) , k+b*np.sin(theta))
plt.title('Random points within Ellipse')
plt.xlabel('x values')
plt.ylabel('y values')

# Generating 100 random points within the Ellipse
r = b*np.random.rand(num_samples)

# Plotting the random points
x, y = h+r*np.cos(theta), k+r*np.sin(theta)
plt.plot(x, y,marker='o')

plt.grid(color='lightgray')
plt.show()
```
