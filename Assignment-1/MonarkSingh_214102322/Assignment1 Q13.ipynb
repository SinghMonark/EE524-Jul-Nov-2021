```python
# Assignment 1 Ques 13

from matplotlib import pyplot as plt

# Defining range for x
x=list(range(-10,11))
y=[]

# Calculating values of y=3x+2
for r in x:
    r=r*3+2
    y.append(r)

# Ploting the given function,giving title and labels
plt.plot(x,y)
plt.title('Graph:y=3x+2')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```
