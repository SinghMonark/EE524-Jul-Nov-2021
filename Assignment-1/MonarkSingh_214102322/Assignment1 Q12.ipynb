```python
#Assignment 1 Ques 12

import numpy as np

# Defining a class for complex numbers

class complex:

# Defining method to find real and imaginary part of complex number
    def __init__(self,real,img):
        self.real=real
        self.img=img
        
# Defining method to print the complex number
    def num(self):
        if int(self.img)>=0:
            print(f"Complex no is: {self.real}+{self.img}j")
        else:
            print(f"Complex no is: {self.real}{self.img}j")

# Defining method to obtain conjugate            
    def conj(self):
        if int(self.img)>=0:
            print(f"\nConjugate of {self.real}+{self.img}j is {self.real}-{self.img}j")
        else:
            print(f"\nConjugate of {self.real}{self.img}j is {self.real}+{abs(self.img)}j ")

# Defining method to obtain absolute value
    def absolute(self):
        import math
        mod=math.sqrt((int(self.real)*int(self.real))+(int(self.img)*int(self.img)))
        if self.img>=0:
            print(f"\nAbsolute value of {self.real}+{self.img}j is {round(mod,2)} ")
        else:
            print(f"\nAbsolute value of {self.real}{self.img}j is {round(mod,2)} ")     

# Defining method to find c1+c2
    def add(self,c):
        if self.img+c[1]>=0:
            print(f"\nc1+c2= {self.real+c[0]}+{self.img+c[1]}j")
        else:
            print(f"\nc1+c2= {self.real+c[0]}{self.img+c[1]}j")

# Defining method to find c1-c2
    def sub(self,c):
        if self.img-c[1]>=0:
            print(f"\nc1-c2= {self.real-c[0]}+{self.img-c[1]}j")
        else:
            print(f"\nc1-c2= {self.real-c[0]}{self.img-c[1]}j")
            
# Defining method to find c1*c2
    def multiply(self,c):
        if (self.real*c[1]+self.img*c[0])>=0:
            print(f"\nc1*c2= {self.real*c[0]-self.img*c[1]}+{self.real*c[1]+self.img*c[0]}j")
        else:
            print(f"\nc1*c2= {self.real*c[0]-self.img*c[1]}{self.real*c[1]+self.img*c[0]}j")

# Defining method to find c1/c2
    def div(self,c):
        n=c[0]*c[0]+c[1]*c[1]
        if (self.img*c[0]-self.real*c[1])>=0:
            print(f"\nc1/c2= {(self.real*c[0]+self.img*c[1])/n}+{(self.img*c[0]-self.real*c[1])/n}j")
        else:
            print(f"\nc1/c2= {(self.real*c[0]+self.img*c[1])/n}{(self.img*c[0]-self.real*c[1])/n}j")

# Defining method to find angle
    def angle(self):
        phi=np.arctan(self.img/self.real)
        if self.img>=0:
            print(f"\nAngle of {self.real}+{self.img}j is {round(phi,2)} rad")
        else:
            print(f"\nAngle of {self.real}{self.img}j is {round(phi,2)} rad")
        
# Defining Complex numbers in a list [Real Part, Img Part]
a=[3,-4]
b=[2,1]

# Definign objects of the class
c1=complex(a[0],a[1])
c2=complex(b[0],b[1])

# Calling the functions
c1.num()
c2.num()
c1.conj()
c2.conj()
c1.add(b)
c1.sub(b)
c1.multiply(b)
c1.div(b)
c1.absolute()
c2.absolute()
c1.angle()
c2.angle()
```
