```python
# Assignment 2 Continuous Distribution Part(b)

# X^2 + Y^2 is an Exponential Distribution

from matplotlib import pyplot as plt
import numpy as np
import math

# Defining a function to plot PDF of Exponential Distribution
def Exponential(L):
    
    y=[]
    domain=np.linspace(0,100,256)
    
    for x in domain:
        value= L * math.exp((-1*L)*x) 
        y.append(value)
    
    Mean=1/L                   # Calculating Mean
    Median= np.log(2)/L        # Calculating Median
    Var=1/(L*L)                # Calculating Variance

# Plotting PDF of Exponential Distribution
    plt.plot(domain,y)
    plt.scatter(Mean,0,c='red', marker='s')
    plt.scatter(Median,0,c='green', marker='^')
    plt.scatter(Var,0,c='yellow',marker='*',s=100)
    plt.legend(['Exponential Dist','Mean','Median','Variance'])
    plt.title('PDF: Exponential Distribution')
    plt.xlabel('x')
    plt.ylabel('f(X=x)')
    plt.grid(True)
    plt.show()

# Calling Exponential function with Lambda = 1/8
Exponential(1/8)
```
