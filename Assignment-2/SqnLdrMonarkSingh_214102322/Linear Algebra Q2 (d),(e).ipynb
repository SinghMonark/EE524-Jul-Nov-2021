```python
'''
Assignment Q2 (d)
 Matrix 1: A=[[1 0]
              [0 1]]
'''
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import math

fig=plt.figure()
fig.set_figwidth(10)
fig.set_figheight(10)
ax=plt.axes(projection='3d')

x1=np.arange(-5,5,0.05)
x2=np.arange(-5,5,0.05)
[x1,x2]=np.meshgrid(x1,x2)

# Calculating the Quadratic form
z= np.square(x1) + np.square(x2)

# Plotting the Surface Plot
ax.plot_surface(x1,x2,z,cmap='viridis')
ax.set_title('3D Surface Plot\n\nQuadratic form: X1^2 + X2^2\n\nHessian Matrix H= [[2 0] [0 2]]')
ax.set_xlabel('X Values')
ax.set_ylabel('Y Values')
ax.set_zlabel('Z Values')
plt.show()

'''
   Eigen Values of the Hessian Matrix are 1=2, 2=2
   Since both Eigen Values are Positive 
   so the quadratic form has a Minima
'''
```

```python
'''
Assignment Q2 (e)
 Matrix : A=[[2 1]
             [1 2]]
'''
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import math

fig=plt.figure()
fig.set_figwidth(10)
fig.set_figheight(10)
ax=plt.axes(projection='3d')

x1=np.arange(-5,5,0.05)
x2=np.arange(-5,5,0.05)
[x1,x2]=np.meshgrid(x1,x2)

# Calculating the Quadratic form
z= 2*np.square(x1) + 2*x1*x2 + 2*np.square(x2)

# Plotting the Surface Plot
ax.plot_surface(x1,x2,z,cmap='viridis')
ax.set_title('3D Surface Plot\n\nQuadratic form: 2*X1^2 + 2*X1*X2 + 2*X2^2 \n\nHessian Matrix H= [[4 2] [2 4]]')
ax.set_xlabel('X Values')
ax.set_ylabel('Y Values')
ax.set_zlabel('Z Values')
plt.show()

'''
   Eigen Values of the Hessian Matrix are 1=2, 2=6
   Since both Eigen Values are Positive it is a Positive Definite Matrix  
   This implies that the Quadratic form is a Convex function  
'''
```
