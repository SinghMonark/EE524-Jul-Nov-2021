```python
# Assignment 2 Discrete Distribution Part(a)
# Bernoulli Distribution

from matplotlib import pyplot as plt

# Define a function to plot PMF for Bernoulli Distribution
def Bernoulli(p):
    
    Mean=p        # Calculating Mean
    Var=p*(1-p)   # Calculating Variance
    
    plt.scatter(Mean,0,c='red', marker='s')
    plt.scatter(Var,0,c='green', marker='^')
    plt.scatter(0,p,c='b', marker='o')
    plt.legend(['Mean','Variance','Probability'])
    plt.scatter(1,1-p,c='b', marker='o')
    plt.xlim(-0.1,1.1)
    plt.ylim(-0.1,1)
    plt.title('PMF: Bernoulli Distribution')
    plt.xlabel('x')
    plt.ylabel('P(X=x)')
    plt.grid(True)
    plt.show()
    
Bernoulli(0.1)
Bernoulli(0.5)
Bernoulli(0.8)
```

```python
# Assignment 2 Discrete Distribution Part(b)

# This is a Binomial Distribution

from matplotlib import pyplot as plt
import math

# Define a function to plot PMF for Binomial Distribution
def Binomial(n,p):
    q=1-p
    y=[]
    domain=list(range(0,21))
    for x in domain:
        c=math.factorial(n)/(math.factorial(n-x)*math.factorial(x))
        value=c*(p**x)*(q**(n-x))
        y.append(value)
    
    Mean=n*p         # Calculating Mean
    Var=n*p*q        # Calculating Variance

    plt.scatter(Mean,0,c='red', marker='s')            # Plotting Mean
    plt.scatter(Var,0,c='green', marker='^',s=100)     # Plotting Variance
    plt.legend(['Mean','Variance'])
    
    for i in domain:
        plt.scatter(i,y[i],c='blue',marker='o')    

    plt.title('PMF: Binomial Distribution')
    plt.xlabel('x')
    plt.ylabel('P(X=x)')
    plt.grid(True)
    plt.show()    
    
# Calling Binomial function with n=20 and p=0.6
Binomial(20,0.6)
```

```python
# Assignment 2 Discrete Distribution Part(c)
# This is a Geometric Distribution

from matplotlib import pyplot as plt

# Define a function to plot PMF for Geometric Distribution
def Geometric(p):
    y=[]
    domain=list(range(1,25))
    for x in domain:
        value=(1-p)**(x-1) * p
        y.append(value)
    
    Mean=1/p            # Calculating Mean
    Var=(1-p)/(p*p)     # Calculating Variance
    
    plt.scatter(Mean,0,c='red', marker='s')      # Plotting Mean
    plt.scatter(Var,0,c='green', marker='^')     # Plotting Variance
    plt.legend(['Mean','Variance'])
    for i in domain:
        plt.scatter(i,y[i-1],c='blue', marker='o')
    
    plt.title('PMF: Geometric Distribution')
    plt.xlabel('x')
    plt.ylabel('P(X=x)')
    plt.grid(True)
    plt.show()
    
# Calling Geometric function with p=0.6
Geometric(0.6)
```

```python
# Assignment 2 Discrete Distribution Part(d)

# This is a Poisson Distribution

from matplotlib import pyplot as plt
import math

# Define a function to plot PMF for Poisson Distribution
def Poisson(m):
    y=[]
    domain=list(range(0,25))
    
    for x in domain:
        value= (math.exp(-1*m) * (m**x)) / math.factorial(x)
        y.append(value)
        
        Mean=m     # Calculating Mean
        Var=m      # Calculating Variance
        
    plt.scatter(Mean,0,c='red', marker='s')
    plt.scatter(Var,0,c='green', marker='^')
    plt.legend(['Mean','Variance'])
    for i in domain:
        plt.scatter(i,y[i],c='blue', marker='o')
    
    plt.title('PMF: Poisson Distribution')
    plt.xlabel('x')
    plt.ylabel('P(X=x)')
    plt.grid(True)
    plt.show()
    
# Calling Poisson function with mean=5
Poisson(5)
```
