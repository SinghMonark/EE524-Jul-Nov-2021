```python
'''
 Assignment Q2 (b)
 Matrix 1: A=[[-1 2]
              [3 4]]
'''
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import math

fig=plt.figure()
fig.set_figwidth(10)
fig.set_figheight(10)
ax=plt.axes(projection='3d')

x1=np.arange(-5,5,0.05)
x2=np.arange(-5,5,0.05)
[x1,x2]=np.meshgrid(x1,x2)

# Calculating the Quadratic form
z= -1*np.square(x1) +5*x1*x2 + 4*np.square(x2) 

# Plotting the Surface Plot
ax.plot_surface(x1,x2,z,cmap='viridis')

ax.set_title('3D Surface Plot\n\nQuadratic Form: -X1^2 + 5*X1*X2 + 4*X2^2\n\nIt is not a valid Covariance Matrix since the quadratic form is not always greater than Zero')
ax.set_xlabel('X Values')
ax.set_ylabel('Y Values')
ax.set_zlabel('Z Values')

plt.show()
```
