```python
# Machine Learning Laboratory
# Assignnment 2
# Continuous Probability Distribution Part(a)

from matplotlib import pyplot as plt
import numpy as np
from sympy import *
import math

# Definig a class for Uniform Distribution

class Uniform:
    
    # Defining method for plotting Probability Density Function (PDF)
    def pdf(self,a,b):
        
        domain=list(range(a,b+1))         # Defining domain of the distribution
        y=[]
        for x in domain:
            value=1/(b-a)                 # Calculating the Prob Density at each x value
            y.append(value)               # Storing the values in y
            
        Mean = (a+b)/2                    # Calculating Mean
        Median= (a+b)/2                   # Calculating Median
        Mode = (a+b)/2                    # Calculating Mode
        Var = (1/12)*np.square(b-a)       # Calculating Variance

        # Plotting the PDF
        plt.plot(domain,y)
        plt.scatter(Mean,0,c='blue',marker='o',s=70)
        plt.scatter(Median,0,c='green',marker='^',s=60)
        plt.scatter(Mode,0,c='yellow',marker='*',s=50)
        plt.scatter(Var,0,c='red',marker='s',s=40)
        plt.xlim(0,b+2)
        plt.ylim(-0.05,(1/(b-a))+0.25)
        plt.title(f'PDF:Uniform Distribution with a= {a},b= {b}')
        plt.xlabel('X')
        plt.ylabel('f(X)')
        plt.legend(['Uniform Dist','Mean','Median','Mode','Variance'])
        plt.grid(True)
        plt.show()
    
    # Defining method for plotting Cumulative Distribution Function (CDF)
    def cdf(self,a,b):    
        y=[]
        domain=list(range(a,b+1))
        for x in domain:
            value=(x-a)/(b-a)
            y.append(value)
        plt.plot(domain,y)
        plt.title(f'CDF:Uniform Distribution with a= {a}, b= {b}')
        plt.xlabel('X')
        plt.ylabel('F(X)')
        plt.legend(['CDF'])
        plt.grid(True)
        plt.show()

# Definig class for Exponential Distribution
class Exponential:

    # Defining class for plotting Probability Density Function (PDF)
    def pdf(self,L):
        y=[]
        domain = np.linspace(0, 10, 128)    # Defining domain of the distribution
        for x in domain:
            value=L*np.exp(-1*L*x)          # Calculating the Prob Density at each x value
            y.append(value)                 # Storing the values in y

        Mean = 1/L                          # Calculating Mean
        Median= np.log(2)/L                 # Calculating Median
        #Mode = mode(L)                     # Calculating Mode
        Var = 1/np.square(L)                # Calculating Variance

        # Plotting Probability Density Function (PDF)
        plt.plot(domain,y)
        plt.scatter(Mean,0,c='blue',marker='o',s=60)
        plt.scatter(Median,0,c='green',marker='^',s=55)
        #plt.scatter(Mode,0,c='yellow',marker='*',s=50)
        plt.scatter(Var,0,c='red',marker='s',s=30)
        plt.title(f'PDF:Exponential Distribution with = {L}')
        plt.xlabel('X')
        plt.ylabel('f(X)')
        plt.legend(['Exp Dist','Mean','Median','Variance'])
        plt.grid(True)
        plt.show()
        
    # Defining method for plotting Cumulative Distribution Function
    def cdf(self,L):    
        y=[]
        domain = np.linspace(0, 10, 128)
        for x in domain:
            value=1-np.exp(-1*L*x)
            y.append(value)
        
        # Plotting CDF
        plt.plot(domain,y)
        plt.title(f'CDF:Exponential Distribution with = {L}')
        plt.xlabel('X')
        plt.ylabel('F(X)')
        plt.legend(['CDF'])
        plt.grid(True)
        plt.show()
    
# Defining class for Gamma Distrbution
class Gamma:
        
    def pdf(self,Y):
        y=[]
        domain = np.linspace(0, 20, 256)
        for x in domain:
            value=((x**(Y-1))*np.exp(-1*x))/(math.gamma(Y))
            y.append(value)

        Mean = Y
        #Median= median(L)
        if Y>1:
            Mode = Y-1
        else:
            Mode = Y 
        Var = Y

        plt.plot(domain,y)
        plt.scatter(Mean,0,c='blue',marker='o',s=80)
        #plt.scatter(Median,0,c='green',marker='^',s=60)
        plt.scatter(Mode,0,c='yellow',marker='*',s=70)
        plt.scatter(Var,0,c='red',marker='s',s=10)
        plt.title(f'PDF:Gamma Distribution with = {Y}')
        plt.xlabel('X')
        plt.ylabel('f(X)')
        plt.legend(['Gamma Dist','Mean','Mode','Variance'])
        plt.grid(True)
        plt.show()
        
    def cdf(self,Y):
        y=[]
        
        domain = np.linspace(0, 20, 128)
        
        for d in domain:
            x = Symbol('x')
            value=integrate((exp(-x)* x**(Y-1))/math.gamma(Y),(x, 0, d))
            y.append(value)
        plt.plot(domain,y)
        plt.title(f'CDF:Gamma Distribution with = {Y}')
        plt.xlabel('X')
        plt.ylabel('F(X)')
        plt.legend(['CDF'])
        plt.grid(True)
        plt.show()
    
# Defining class for Beta Distribution
class Beta:
        
    def pdf(self,a,b):
        B=(math.gamma(a)*math.gamma(b))/math.gamma(a+b)
        y=[]
        domain = np.linspace(0, 1, 128)
        for x in domain:
            value=((x**(a-1))*(1-x)**(b-1))/B
            y.append(value)

        Mean = a/(a+b)
        Median= (a-0.33)/(a+b-0.67)
        Mode = (a-1)/(a+b-2)
        Var = (a*b)/(np.square(a+b)*(a+b+1))

        plt.plot(domain,y)
        plt.scatter(Mean,0,c='blue',marker='o',s=80)
        plt.scatter(Median,0,c='green',marker='^',s=60)
        plt.scatter(Mode,0,c='yellow',marker='*',s=60)
        plt.scatter(Var,0,c='red',marker='s',s=40)
        plt.title(f'PDF:Beta Distribution with = {a}, = {b}')
        plt.xlabel('X')
        plt.ylabel('f(X)')
        plt.legend(['Beta Dist','Mean','Median','Mode','Variance'])
        plt.grid(True)
        plt.show()
        
    def cdf(self,a,b):
        B=(math.gamma(a)*math.gamma(b))/math.gamma(a+b)
        y=[]
        domain = np.linspace(0, 1, 128)
        
        for d in domain:
            x = Symbol('x')
            value=integrate(((x**(a-1))*(1-x)**(b-1))/B,(x, 0, d))
            y.append(value)
        plt.plot(domain,y)
        plt.title(f'CDF:Beta Distribution with = {a}, = {b}')
        plt.xlabel('X')
        plt.ylabel('F(X)')
        plt.legend(['CDF'])
        plt.grid(True)
        plt.show()
    
# Defining class for Gaussian Distribution
class Gaussian:
         
    def pdf(self,m,s):
        y=[]
        c=1/(s*math.sqrt(2*np.pi))
        domain = np.linspace(m-6*s,m+6*s, 64)
        for x in domain:
            value=c*np.exp(-0.5*np.square((x-m)/s))
            y.append(value)

        Mean = m
        Median= m
        Mode = m
        Var = np.square(s)

        plt.plot(domain,y)
        plt.scatter(Mean,0,c='blue',marker='o',s=80)
        plt.scatter(Median,0,c='green',marker='^',s=60)
        plt.scatter(Mode,0,c='yellow',marker='*',s=50)
        plt.scatter(Var,0,c='red',marker='s',s=40)
        plt.title(f'PDF:Gaussian Distribution with = {m}, = {s}')
        plt.xlabel('X')
        plt.ylabel('f(X)')
        plt.legend(['Gaussian Dist','Mean','Median','Mode','Variance'])
        plt.grid(True)
        plt.show()
        
    def cdf(self,m,s):
        y=[]
        domain = np.linspace(m-6*s,m+6*s, 64)
        
        for x in domain:
            value=0.5*(1+math.erf((x-m)/(s*math.sqrt(2))))
            y.append(value)
        plt.plot(domain,y)
        plt.title(f'CDF:Gaussian Distribution with = {m}, = {s}')
        plt.xlabel('X')
        plt.ylabel('F(X)')
        plt.legend(['CDF'])
        plt.grid(True)
        plt.show()

# Defining Standard Normal Distribution with mean=0, standard deviation=1
class Std_Normal:
    def pdf(self):
        m=0
        s=1
        y=[]
        c=1/(s*math.sqrt(2*np.pi))
        domain = np.linspace(m-6*s,m+6*s, 64)
        for x in domain:
            value=c*np.exp(-0.5*np.square((x-m)/s))
            y.append(value)

        Mean = m
        Median= m
        Mode = m
        Var = np.square(s)

        plt.plot(domain,y)
        plt.scatter(Mean,0,c='blue',marker='o',s=80)
        plt.scatter(Median,0,c='green',marker='^',s=60)
        plt.scatter(Mode,0,c='yellow',marker='*',s=50)
        plt.scatter(Var,0,c='red',marker='s',s=40)
        plt.title(f'PDF:Standard Normal Distribution with =0, =1')
        plt.xlabel('X')
        plt.ylabel('f(X)')
        plt.legend(['Gaussian Dist','Mean','Median','Mode','Variance'])
        plt.grid(True)
        plt.show()
        
    def cdf(self):
        m=0
        s=1
        y=[]
        domain = np.linspace(m-6*s,m+6*s, 64)
        
        for x in domain:
            value=0.5*(1+math.erf((x-m)/(s*math.sqrt(2))))
            y.append(value)
        plt.plot(domain,y)
        plt.title(f'CDF:Gaussian Distribution with = 0, = 1')
        plt.xlabel('X')
        plt.ylabel('F(X)')
        plt.legend(['CDF'])
        plt.grid(True)
        plt.show()
    
# Defining class for Chi-squared distributionclass Chi_squared:
class Chi_squared:        
    def pdf(self,k):
        y=[]
        domain = np.linspace(0, 15, 64)
        
        for x in domain:
            value=(1/(2**(k/2)*math.gamma(k/2)))*(x**((k/2)-1)*math.exp(-0.5*x))
            y.append(value)

        Mean = k
        Median= k*(1-(2/(9*k)))**3
        if k-2>=0:
            Mode = k-2
        else:
            Mode=0
        Var = 2*k

        plt.plot(domain,y)
        plt.scatter(Mean,0,c='blue',marker='o',s=80)
        plt.scatter(Median,0,c='green',marker='^',s=60)
        plt.scatter(Mode,0,c='yellow',marker='*',s=100)
        plt.scatter(Var,0,c='red',marker='s',s=50)
        plt.title(f'PDF:Chi-Squared Distribution with k= {k}')
        plt.xlabel('X')
        plt.ylabel('f(X)')
        plt.legend(['Chi-Squared Dist','Mean','Median','Mode','Variance'])
        plt.grid(True)
        plt.show()
        
    def cdf(self,k):
        y=[]
        domain = np.linspace(0, 15, 64)
        
        for d in domain:
            x = Symbol('x')
            value=integrate((1/(2**(k/2) * math.gamma(k/2))) * x**((k/2)-1) * exp(-0.5*x),(x, 0, d))
            y.append(value)
            
        plt.plot(domain,y)
        plt.title(f'CDF:Chi-Squared Distribution with k= {k}')
        plt.xlabel('X')
        plt.ylabel('F(X)')
        plt.legend(['CDF'])
        plt.grid(True)
        plt.show()
    
# Calling all Probability Distributions

uni=Uniform()
uni.pdf(2,10)
uni.pdf(3,7)
uni.pdf(1,5)
uni.cdf(1,5)

expo=Exponential()
expo.pdf(0.5)
expo.pdf(1)
expo.pdf(5)
expo.cdf(1)

gam=Gamma()
gam.pdf(1)
gam.pdf(3)
gam.pdf(9)
gam.cdf(1)

bet=Beta()
bet.pdf(0.5,0.5)
bet.pdf(5,1)
bet.pdf(2,5)
bet.cdf(2,5)

gauss=Gaussian()
gauss.pdf(0,4)
gauss.pdf(25,15)
gauss.pdf(-15,9)
gauss.cdf(0,2)

stdn=Std_Normal()
stdn.pdf()
stdn.cdf()

chi=Chi_squared()
chi.pdf(1)
chi.pdf(3)
chi.pdf(6)
chi.cdf(1)
```
