```python
'''
 Assignment2 Q2 (a)
 Matrix 1: A=[[0.5 0]
              [0 0.5]]
'''
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import math

fig=plt.figure()
fig.set_figwidth(10)
fig.set_figheight(10)
ax=plt.axes(projection='3d')

x1=np.arange(-1.5,1.5,0.05)
x2=np.arange(-1.5,1.5,0.05)
[x1,x2]=np.meshgrid(x1,x2)

 # Calculating the Quadratic form
z=0.5*np.square(x1)+0.5*np.square(x2)   

# Defining the Eigen Vectors
X1,Y1,Z1=[0,0,0]
X2,Y2,Z2=[0,0,0]
U1,V1,W1=[1,0,0]
U2,V2,W2=[0,1,0]

# Plotting the Eigen Vectors
plt.quiver(X1, Y1, Z1, U1, V1, W1, color='b',label='Eigen Vector V1 corresponding to 1')
plt.quiver(X2, Y2, Z2, U2, V2, W2, color='r',label='Eigen Vector V2 corresponding to 2')

# Plotting the extent {sqrt(Eigen Value)} along the Eigen Vector 
ax.scatter3D(math.sqrt(0.5),0,0,c='purple',marker='*',s=100,label='Extent along Eigen Vector V1')
ax.scatter3D(0,math.sqrt(0.5),0,c='green',marker='*',s=100,label='Extent along Eigen Vector V2')

# Plotting the contour plot of the Quadratic form
ax.contour(x1,x2,z,20)
ax.set_title('Contour Plot Top View\nQuadratic Form: 0.5*X1^2 + 0.5*X2^2\nEigen Values: 1=0.5, 2=0.5 ')
ax.set_xlabel('X1 Values')
ax.set_ylabel('X2 Values')
ax.set_zlabel('Z Values')

ax.legend()
# Setting azimuth and elevation for Top View
ax.view_init(azim=-90,elev=90)      
plt.show()
```

```python
'''
 Assignment2 Q2 (a)
 Matrix 6: A=[[9 5]
              [5 4]]
'''
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import math

fig=plt.figure()
fig.set_figwidth(20)
fig.set_figheight(20)

x1=np.arange(-6,6,0.1)
x2=np.arange(-6,6,0.1)

[x1,x2]=np.meshgrid(x1,x2)

# Calculating the Quadratic form
z= 9*np.square(x1) + 10*x1*x2 + 4*np.square(x2)  

# Defining the Eigen Vectors
X1,Y1,Z1=[0,0,0]
X2,Y2,Z2=[0,0,0]
U1,V1,W1=[1.618,1,0]
U2,V2,W2=[1,-1.618,0]

# Computing the extent along first Eigen Vector (V1)
p1=1
b1=1.618
h1=math.sqrt(p1*p1+b1*b1)
sin_theta1=p1/h1
cos_theta1=b1/h1
extent_x1x=math.sqrt(12.09)*cos_theta1        # X coordinate along Eigen Vector (V1)
extent_x1y=math.sqrt(12.09)*sin_theta1        # Y coordinate along Eigen Vector (V2)

# Computing the extent along second Eigen Vector (V2)
p2=-1.618
b2=1
h2=math.sqrt(p2*p2+b2*b2)
sin_theta2=p2/h2
cos_theta2=b2/h2
extent_x2x=math.sqrt(0.91)*cos_theta2       # X coordinate along Eigen Vector (V1)
extent_x2y=math.sqrt(0.91)*sin_theta2       # Y coordinate along Eigen Vector (V2)

# Transforming the orthogonal axes along the Eigen Vectors
x= x1*cos_theta1 + x2*sin_theta1
y=-1*x1*sin_theta1 + x2*cos_theta1

# Plotting the Eigen Vectors
ax = fig.add_subplot(2,1,1, projection='3d')
ax.quiver(X1, Y1, Z1, U1, V1, W1, color='b',label='Eigen Vector V1 corresponding to 1')
ax.quiver(X2, Y2, Z2, U2, V2, W2, color='r',label='Eigen Vector V2 corresponding to 2')

# Plotting the extent {sqrt(Eigen Value)} along the Eigen Vector
ax.scatter3D(extent_x1x,extent_x1y,0,c='purple',marker='*',s=100,label='Extent along Eigen Vector V1')
ax.scatter3D(extent_x2x,extent_x2y,0,c='green',marker='*',s=100,label='Extent along Eigen Vector V2')

# Plotting the contour plot of the Quadratic form
ax.contour3D(x1,x2,z,20)
ax.set_title('Contour Plot Top View\nQuadratic Form: 9*X1^2 + 10*x1*x2 + 4*X2^2\nEigen Values: 1=12.09, 2=0.91')
ax.set_xlabel('X1 Values')
ax.set_ylabel('X2 Values')
ax.set_zlabel('Z Values')
ax.legend()
ax.view_init(azim=-90,elev=90)

# Plotting the contour plot with Eigen Vectors as the Orthogonal Axes
ax = fig.add_subplot(2,1,2, projection='3d')
ax.contour3D(x,y,z,20)
ax.set_title('Contour Plot Top View \nwith Eigen Vectors as Orthogonal Axes of the plot')
ax.set_xlabel('X1 Values')
ax.set_ylabel('X2 Values')
ax.set_zlabel('Z Values')
ax.view_init(azim=-90,elev=90)
plt.show()
```
